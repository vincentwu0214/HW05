```python
import requests
```

```python
url = "http://finance.google.com/finance/historical"
```

```python
params = {}
params['q'] = 'GOOG'
params['startdate'] = '2000-01-01'
params['output'] = 'csv'
```

```python
r = requests.get(url, params=params)
```

```python
print(r.text)
```
